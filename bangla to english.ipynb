```python
from transformers import MarianMTModel, MarianTokenizer
import pandas as pd

# মডেলের নাম (বাংলা থেকে ইংরেজি অনুবাদ)
MODEL_NAME = "Helsinki-NLP/opus-mt-bn-en"

# মডেল এবং টোকেনাইজার লোড করা
tokenizer = MarianTokenizer.from_pretrained(MODEL_NAME)
model = MarianMTModel.from_pretrained(MODEL_NAME)

def translate_text(text):
    """টেক্সটকে ইংরেজিতে অনুবাদ করে"""
    if not isinstance(text, str) or text.strip() == "":
        return text
    inputs = tokenizer.encode(text, return_tensors="pt", truncation=True, max_length=512)
    translated = model.generate(inputs, max_length=512, num_beams=5, early_stopping=True)
    return tokenizer.decode(translated[0], skip_special_tokens=True)

# CSV ফাইল লোড করা
data = pd.read_csv("bangla_data.csv")

# প্রতিটি কলামে অনুবাদ প্রয়োগ করা
for col in data.columns:
    print(f"Translating column: {col}")
    data[col] = data[col].apply(translate_text)

# নতুন ফাইলে সেভ করা
data.to_csv("translated_data.csv", index=False)
print("Translation complete. File saved as 'translated_data.csv'")
```
